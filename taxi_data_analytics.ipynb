```python
def get_hour(datetime):
    return int(datetime.split(' ')[1].split(':')[0])

def get_minute(datetime):
    return int(datetime.split(' ')[1].split(':')[1])

def duration(pickup, dropoff):
    if pickup.split(' ')[0] != dropoff.split(' ')[0]:
        duration = 60*(23-get_hour(pickup))+59-get_minute(pickup) # time till end of the day
        duration += 60*(get_hour(dropoff))+get_minute(dropoff) # time after midnnight
    else:
        duration = 60*(get_hour(dropoff)-get_hour(pickup)) + get_minute(dropoff) - get_minute(pickup)
    return duration

# For trips through midnight consider a day of pickup as a day of trip because of suggestion of fares formation at start of a trip
def get_weekday(datetime):
    date = datetime.split(' ')[0]
    if date.split('-')[1] == '05':
        return 'Friday'
    days = {0: 'Friday', 1: 'Saturday', 2: 'Sunday', 3: 'Monday', 4: 'Tuesday', 5: 'Wednesday', 6: 'Thursday'}
    return days[int(date.split('-')[2])%7]

def get_boroughs(pickup,dropoff,lookup):
    output = [lookup['Borough'][pickup], lookup['Borough'][dropoff]]
    output.sort()
    return '-'.join(output)
```

```python
taxi['trip_duration'] = taxi.apply(lambda x: duration(x['tpep_pickup_datetime'], x['tpep_dropoff_datetime']), axis = 1)
taxi['day_of_week'] = taxi['tpep_pickup_datetime'].apply(get_weekday)
taxi.head()
```

```python
taxi = taxi.loc[taxi['passenger_count'] != 0]
taxi = taxi.loc[taxi['trip_distance'] != 0.0]
taxi = taxi.loc[taxi['trip_duration'] != 0]
```
